```python
# Criando cópia de arquivo na pasta criada

import shutil

arquivo_copiar = Path('Arquivos_Lojas/202001_Morumbi_SP.csv')
arquivo_colar = Path('Pasta Auxiliar/202001_Morumbi_SP_versao2.csv')

shutil.copy2(arquivo_copiar, arquivo_colar)
```

```python
# Separar os arquivos conforme o nome do estado que apresentam

def extrair_estado(path):
    return path.name.split('.csv')[0][-2:]


def checar_existencia_pasta(path):
    if not path.exists():
        path.mkdir(parents=True)        


def mover_arquivo(arquivo, dst):
    shutil.move(Path(arquivo), dst)


def mover_arquivos(nomePasta):
    caminho = Path(nomePasta)
    arquivos = caminho.iterdir()

    for arquivo in arquivos:
        estado = extrair_estado(arquivo)

        path = Path(f'Arquivos_Lojas_Estado/{estado}')

        checar_existencia_pasta(path)
        mover_arquivo(arquivo, path)


mover_arquivos('Arquivos_Lojas')
    
```
